## **Test**

```
from imageai.Prediction import ImagePrediction
import os
from IPython.display import Image

execution_path = os.getcwd()

prediction = ImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, "resnet50_weights_tf_dim_ordering_tf_kernels.h5"))
prediction.loadModel()

!wget -O object.jpg https://thanhlammedia.vn/wp-content/uploads/2019/09/iphone.jpg
object_prediction = "object.jpg"

predictions, probabilities = prediction.predictImage(os.path.join(execution_path, object_prediction), result_count=5 )
for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)

display(Image(object_prediction))
```

## **Test**

```


from imageai.Detection import ObjectDetection
import os
from IPython.display import Image

execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "yolo.h5"))
detector.loadModel()


!wget -O object.jpg https://product.hstatic.net/1000360516/product/ts001__1___copy__730da795f9b64b05a1739a7699e33974_grande.jpg
object_prediction = "object.jpg"

!wget -O Object_Prediction_Output.jpg https://drive.google.com/file/d/1MhcKfoNdHgU_5jhzkuyWlJoJIy8r6q1h/view?usp=sharing
Object_Prediction_Output = "Object_Prediction_Output.jpg"

detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , object_prediction), output_image_path=os.path.join("" , Object_Prediction_Output), minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")

display(Image(Object_Prediction_Output))
```
