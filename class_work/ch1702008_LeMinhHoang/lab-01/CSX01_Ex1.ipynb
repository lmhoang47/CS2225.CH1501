```
import math 

a = float(input())
b = float(input())
c = float(input())

s = (a + b + c) / 2

area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

if int(area) == float(area):
   area = int(area)
else:
   area = float(area)

if not(a+b > c and b+c > a and a+c > b):
  print("Khong phai tam giac")
elif c == math.sqrt(a**2 + b**2) or b == math.sqrt(a**2 + c**2) or a == math.sqrt(c**2 + b**2):
  print("Tam giac vuong, dien tich = {0:.2f}".format(area).rstrip('0').rstrip('.'))
elif a==b and b==c:
  print("Tam giac deu, dien tich = {0:.2f}".format(area).rstrip('0').rstrip('.'))
elif a==b or b==c or c==a:
  print("Tam giac can, dien tich = {0:.2f}".format(area).rstrip('0').rstrip('.'))
else:
  print("Tam giac thuong, dien tich = {0:.2f}".format(area).rstrip('0').rstrip('.'))
```

```
n = int(input())
k = int(input())
p = int(input())
q = int(input())

position = 2*(p-1) + q

prev_position = position - k
after_position = position + k

if prev_position >= 1:
    print((prev_position-1)//2 + 1, (prev_position-1)%2 + 1)
elif after_position <= n:
    print((after_position-1)//2 + 1, (after_position-1)%2 + 1)
else:
    print(-1)
```

```
x = int(input())

n1, n2 = 0, 1
count = 0


if not(x >= 1 and x <= 30):
  print("So {n} khong nam trong khoang [1,30].".format(n=x))
else:
  while count < x:
       nth = n1 + n2
       n1 = n2
       n2 = nth
       count += 1

       if count >= x:
         print(n1)
```

```
n = str(input())
count = 0
lucky = ["4", "7"]

for char in n:
  if char in lucky:
    count += 1

if count == 4 or count == 7:
  print("YES")
else:
  print("NO")
```
