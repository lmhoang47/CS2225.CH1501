```
from re import search

def check_suffix(list_string):
  for x in list_string:
    if not(str(x).endswith('lios') == True or str(x).endswith('liala') == True  or str(x).endswith('etr') == True  or str(x).endswith('etra') == True  or str(x).endswith('initis') == True  or str(x).endswith('inites') == True):
      return 0
  return 1

def check_gender(list_string):
  flag_male = 0
  flag_female = 0
  for x in list_string:
    if str(x).endswith('lios') == True or str(x).endswith('etr') == True or str(x).endswith('initis') == True:
      flag_male = 1
    else:
      flag_female = 1
    if flag_male == 1 and flag_female == 1:
      return 0
  return 1

def check_nounfrq(list_string):
  count = 0
  for x in list_string:
    if  str(x).endswith('etr') == True or  str(x).endswith('etra') == True:
      count += 1
  if count > 1 or count == 0 :
    return 0
  else:
    return 1

def check_ordinal(list_string):
  post_noun = []
  post_adj = []
  post_verb = []
  
  for index, item in enumerate(list_string):
    if str(item).endswith('etr') == True or str(item).endswith('etra') == True:
      post_noun.append(index)
    if str(item).endswith('lios') == True or str(item).endswith('liala') == True:
      post_adj.append(index)
    if str(item).endswith('initis') == True or str(item).endswith('inites') == True:
      post_verb.append(index)

  if (post_adj and max(post_adj) > min(post_noun)) or (post_verb and min(post_verb) < max(post_noun)):
    return 0
  return 1

list_input = list(map(str,input().strip().split(' ')))

while 1:
  if check_suffix(list_input) == 0:
    print('NO')
    break
  if len(list_input) == 1:
    print('YES')
    break
  if check_gender(list_input) == 0:
    print('NO')
    break
  if check_nounfrq(list_input) == 0:
    print('NO')
    break
  if check_ordinal(list_input) == 0:
    print('NO')
    break

  print('YES')
  break
```
