```
x,y = map(int, input().split())

arrays = []

#Up
step = 0
i = 1
j = 0

while i < x - step:
  while j < y - step:
    if (i,j) not in arrays:
      arrays.append((i,j))
    j += 1
  i += 1
  j = 1
  step += 1

#Right
step = 0
i = 1
j = y-1

while j < y - step and j > 1:
  while i < x - step:
    if (i,j) not in arrays:
      arrays.append((i,j))
    i += 1
  i = 2
  j -= 1
  step += 1

#Down
step = 0
i = x-1
j = y-1

while i < x - step and i > 1:
  while j < y - step and j > 1:
    if (i,j) not in arrays:
      arrays.append((i,j))
    j -= 1
  i -= 1
  j = y-2
  step += 1

#Left
step = 0
i = x-2
j = 1

while j < y - step:
  while i < x - step and i > 1:
    if (i,j) not in arrays:
      arrays.append((i,j))
    i -= 1
  i = x-2
  j += 1
  step += 1

print(*arrays)
print(len(arrays))
```

```
from math import ceil

n, m, a = map(int, input().split())

if n <= a and m <= a:
  print(1)
elif n <= a and m > a:
  print(ceil(m/a))
elif n > a and m <= a:
  print(ceil(n/a))
else:
  print(ceil(n/a) * ceil(m/a))
```

```
import math

x = int(input())
list_output = []

for num in range(3,x,2):
  if all(num%i!=0 for i in range(3,int(math.sqrt(num))+1, 2)):
    list_output.append(num)

if x in (4,5):
  print(2)
else:
  print(list_output[len(list_output)-2])
```

```
num = int(input())

order = len(str(num))

sum = 0

temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** order
   temp //= 10

if num == sum:
   print("True")
else:
   print("False")
```
